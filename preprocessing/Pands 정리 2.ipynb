### 1. groupby

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby("Team")["Points"].sum()
```

```python
h_index = df.groupby(["Team", "Year"]).sum()
h_index
```

```python
h_index.index
```

```python
h_index["Devils" : "Kings"]
```

```python
h_index.unstack() #unstack : 그룹으로 묶여진 데이터를 matrix 형태로 전환
```

```python
h_index.swaplevel() #index level변경
```

```python
grouped= df.groupby("Team")
```

```python
grouped.get_group("Riders") #Team으로 그룹시킨것들 중 "Riders"만
```

```python
df["Team"].unique()
```

#### agg : aggregation key값 별로 요약된 정보

```python
grouped.agg(min)
```

```python
import numpy as np
grouped.agg(np.mean)
```

```python
grouped["Points"].agg([np.sum, np.mean, np.std])
```

```python
nlz = lambda x: (x - x.mean()) / x.std()
grouped.transform(nlz)
```

```python
import dateutil
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.groupby('month')['duration'].sum() #month별 duration 합
```

```python
df_phone[df_phone["item"]=='call'].groupby('month')['duration'].sum()
```

```python
df_phone.groupby(['month','item'])['date'].count().unstack()
```

```python
df_phone.groupby('month', as_index=False).agg({"duration" : "sum"})
```

```python
df_phone.groupby('month', as_index=False)["duration"].sum()
```

```python
df_phone.groupby(['month','item']).agg({'duration' : 'sum',
                                       'network_type' : 'count',
                                       'date' : 'first'})
#column 별로 다른 연산 가능
```

```python
df_phone.groupby(['month', 'item']).agg({'duration' : 'min',
                                        'network_type' : 'count',
                                        'date' : ['min','first','nunique']})
#한 column 당 여러 연산 가능. 
```

```python
grouped = df_phone.groupby('month').agg({'duration' : ['min','max','mean']})
grouped
```

```python
grouped.columns
```

```python
grouped.columns = grouped.columns.droplevel(level=0)
grouped.columns
```

```python
grouped.index
```

```python
grouped.rename(columns = {'min' : 'min_duration', 'max' : 'max_duration',
                         'mean' : 'mean_duration'})
```

### 3. merge, concat

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data)
df_b
```

```python
pd.merge(df_a, df_b, on="subject_id") #subject_id가 같은 것끼리만
```

```python
pd.merge(df_a, df_b, left_on = 'subject_id',right_on = 'subject_id')
#column name이 다를 때 적용
```

```python
pd.merge(df_a, df_b, on = 'subject_id', how = 'left') #왼쪽기준
```

```python
pd.merge(df_a, df_b, on = 'subject_id', how = 'right') #오른쪽기준
```

```python
pd.merge(df_a, df_b, on = 'subject_id', how = 'outer') #all
```

```python
pd.merge(df_a, df_b, right_index=True, left_index=True)
#둘 다 살려서 억지로 끼워맞춤
```
